### ðŸ§¹ Data Cleaning Insight
Cleaned and standardized order data, calculated new business metrics (Discount, Sale Price, Profit),
and prepared the dataset for analysis by handling missing values and formatting dates.

```python
import pandas as pd
```

```python
import pandas as pd
df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.head()
```

```python
df[df['ship_mode'].isnull()]
```

```python
df['ship_mode'] = df['ship_mode'].fillna(df['ship_mode'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Discount 
df['Discount']=df['list_price']*df['discount_percent']*.01
```

```python
df
```

```python
# sale Price
df['Sale_Price']=df['list_price']-df['Discount']
```

```python
# profit 
df['Profit']=df['Sale_Price']-df['cost_price']
```

```python
df
```

```python
# convert order_date object to datetime format
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
df
```

```python
# drop the columns
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```
